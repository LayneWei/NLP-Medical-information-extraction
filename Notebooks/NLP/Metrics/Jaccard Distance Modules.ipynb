```python
import pandas as pd
import numpy as np
import nltk
import sklearn
import os
import csv
from nltk.stem import WordNetLemmatizer
```

```python
def extract_txt(filename): #build dictionary of words count percentage from all 18 files
    all_txt = []
    with open(filename) as csvfile:
        csvdata = csv.reader(csvfile)
        next(csvdata, None)
        for row in csvdata:
            all_txt.append([row[13], row[3]])
    return all_txt


all_txt = extract_txt('combined_diseases.csv')
print(all_txt)
```

```python
lemmatizer = WordNetLemmatizer() 
words_tokens = []
for j in range(0,len(all_txt)):
    tokenized = nltk.word_tokenize(all_txt[j][0])
    tokenized_lem = []
    for n in tokenized:
        tokenized_lem.append(lemmatizer.lemmatize(n).lower())
    tokenized_2 = nltk.word_tokenize(all_txt[j][1])
    tokenized_lem2 = []
    for n in tokenized_2:
        tokenized_lem2.append(lemmatizer.lemmatize(n).lower())
    words_tokens.append([tokenized_lem, tokenized_lem2])
    

for words in words_tokens:
    print('First 2 tokens of range txt: ', words[0][:2])
    print('First 2 tokens of canon txt: ', words[1][:2])
```

```python
from numpy import argmax

def transform(headlines):
    tokens = [w for s in headlines for w in s ]
    print()
    print('All Tokens:')
    print(tokens)

    results = []
    label_enc = sklearn.preprocessing.LabelEncoder()
    onehot_enc = sklearn.preprocessing.OneHotEncoder()
    
    encoded_all_tokens = label_enc.fit_transform(list(set(tokens)))
    encoded_all_tokens = encoded_all_tokens.reshape(len(encoded_all_tokens), 1)
    
    onehot_enc.fit(encoded_all_tokens)
    
    for headline_tokens in headlines:
        print()
        print('Original Input:', headline_tokens)
        
        encoded_words = label_enc.transform(headline_tokens)
        print('Encoded by Label Encoder:', encoded_words)
        
        encoded_words = onehot_enc.transform(encoded_words.reshape(len(encoded_words), 1))
        print('Encoded by OneHot Encoder:')
        print(encoded_words)

        results.append(np.sum(encoded_words.toarray(), axis=0))
    
    return results


tranformed_results = []
for concept in words_tokens:
    transformed_results.append(transform(concept))
    #transformed_results.append(transform(words[0]))
    
print(transformed_results[5])
```

```python
def calculate_position(values):
    x = []
    for pos, matrix in enumerate(values):
        if matrix > 0:
            x.append(pos)
    return x

"""
    Since scikit-learn can only compare same number of dimension of input. 
    Add padding to the shortest sentence.
"""
def padding(sentence1, sentence2):
    x1 = sentence1.copy()
    x2 = sentence2.copy()
    
    diff = len(x1) - len(x2)
    
    if diff > 0:
        for i in range(0, diff):
            x2.append(-1)
    elif diff < 0:
        for i in range(0, abs(diff)):
            x1.append(-1)
    
    return x1, x2    

y_actual = calculate_position(transformed_results[0])

for j in range(3,len(words_tokens)):
    print('Range Txt: %s' % words_tokens[j][0])
    for i, m in enumerate(words_tokens[j]):
        y_compare = calculate_position(transformed_results[j][i])
        x1, x2 = padding(y_actual, y_compare)
        score = sklearn.metrics.jaccard_similarity_score(x1, x2)
        print('-----')
        print('Score: %.2f, Comparing Sentence: %s' % (score, words_tokens[j][i]))
```
