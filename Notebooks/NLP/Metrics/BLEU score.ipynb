```python
import pandas as pd
import numpy as np
import nltk
import sklearn
import os
import csv
from nltk.stem import WordNetLemmatizer
from nltk.translate.bleu_score import sentence_bleu
```

```python
def extract_txt(filename): #build dictionary of words count percentage from all files
    all_txt = []
    with open(filename) as csvfile:
        csvdata = csv.reader(csvfile)
        next(csvdata, None)
        for row in csvdata:
            all_txt.append([row[13], row[3]])
    return all_txt

all_txt = extract_txt('combined_diseases2.csv')
print(all_txt)
```

```python
lemmatizer = WordNetLemmatizer() 
words_tokens = []
for j in range(0,len(all_txt)):
    tokenized = nltk.word_tokenize(all_txt[j][0])
    tokenized_lem = []
    for n in tokenized:
        tokenized_lem.append(lemmatizer.lemmatize(n).lower())
    tokenized_2 = nltk.word_tokenize(all_txt[j][1])
    tokenized_lem2 = []
    for n in tokenized_2:
        tokenized_lem2.append(lemmatizer.lemmatize(n).lower())
    words_tokens.append([tokenized_lem, tokenized_lem2])
    
for words in words_tokens:
    print('tokens of range txt: ', words[0][:])
    print('tokens of canon txt: ', words[1][:])
```

```python
references = [] #canon text
candidates = [] #range text

for n in range(len(words_tokens)):
    references.append(words_tokens[n][1])
for n in range(len(words_tokens)):
    candidates.append(words_tokens[n][0])
for n in references: #canon text
    print ('reference :{}'.format(n))
for n in candidates: #range text
    print ('candidate :{}'.format(n))
```

```python
for i in range(0, len(candidates)):
    reference_list = []
    reference_list.append(references[i])
    print (candidates[i], reference_list)
    print('BLEU: ', sentence_bleu(reference_list, candidates[i], weights = (1, 0, 0, 0)))
```
